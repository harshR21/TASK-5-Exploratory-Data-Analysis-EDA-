```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\Harsh\Downloads\titanic_sample.csv")

#STEP 1 a.Use .describe(), .info(), .value_counts() --

data.describe()
data.info()
data['Survived'].value_counts()

#STEP 2 b.Use sns.pairplot(), sns.heatmap() for visualization

numeric_cols = ["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]

#HeatMap
sns.heatmap(data
            [numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Titanic Features")
plt.show()
 
#PairPlot
#sns.pairplot(data           
 #           [numeric_cols, hue == 'Survived'])

sns.pairplot(data[numeric_cols], hue="Survived")
plt.suptitle("Pairplot of Titanic Features by Survival", y=1.02)
plt.show()

#c.Identify relationships and trends
'''- Females survived at a higher rate than males.  
   -Fare tends to be higher for survivors  suggests wealth/class mattered.
   -Younger passengers (children) have slightly higher survival probability. '''

#d.Plot histograms, boxplots, scatterplots

 #Histogram of Age
data["Age"].hist(bins=20, edgecolor="black")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

 #Boxplot of Age VS Survived
sns.boxplot(x='Survived' , y='Age', data=data)
plt.title('Age Distribution by Survival')
plt.show()

 #Scatterplot Age vs Fare 
sns.scatterplot(x='Age' , y='Fare' , data=data)
plt.title("Age VS Fare colored by Survival")
plt.show()

#(e) Write Observations for Each Visual

 #Age Histogram:
print("Most passengers are between 20-40 years. Fewer very young or very old passengers.")

 #Age vs Survived Boxplot:
print("Survivors had slightly lower median age, with several young children among them. Outliers exist in both groups.")

 #Age vs Fare Scatterplot:
print("Passengers who paid higher fares (top right) are mostly survivors and likely from 1st class.Most non-survivors had low fare and lower age.")

#F Summary of findings

##Summary of Titanic EDA:

print("""
1. Most passengers were in 3rd class and paid low fares.
2. Females and 1st class passengers had much higher survival rates.
3. Age has a wide distribution; younger passengers slightly more likely to survive.
4. Fare correlates positively with survival  those who paid more tended to survive.
5. Pclass (1st > 2nd > 3rd) strongly influences survival; many 3rd class passengers did not survive.
6. The correlation heatmap shows few strong correlations; categorical variables like Sex are critical.

Conclusion:
Survival was influenced more by **class**, **sex**, and **fare paid** than just numeric variables like age or family size.
"""
 )
```
